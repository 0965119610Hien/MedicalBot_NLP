```python
import PyPDF2
from PyPDF2 import PdfReader
pdf = open(f'E:\\NLP\\NaturalLanguageProcess\\Caythuoc.pdf', 'rb')
pdf_reader = PdfReader(pdf)
print(f'The PDF has {len(pdf_reader.pages)} pages.')
print('Info: ', pdf_reader.metadata)
```

```python
import fitz  
import pytesseract
from PIL import Image
import io

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

pdf_document = fitz.open('E:\\NLP\\NaturalLanguageProcess\\Caythuoc.pdf')

for page_num in range(len(pdf_document)):
    page = pdf_document[page_num]
    pix = page.get_pixmap(matrix=fitz.Matrix(2, 2)) 
    img = Image.open(io.BytesIO(pix.tobytes()))
    
    text = pytesseract.image_to_string(img, lang='vie')  
    print(f"--- Trang {page_num+1} ---")
    print(text)
    print()

pdf_document.close()
```
