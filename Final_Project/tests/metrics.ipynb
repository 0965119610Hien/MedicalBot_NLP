```python
data_revised = pd.read_csv('medical_eval_report_revised.csv')
data_revised.head()
```

```python
metrics = ['faithfulness', 'answer_relevancy', 'answer_correctness', 'context_precision']
averages = data_revised[metrics].mean()
print(averages)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập phong cách cho biểu đồ
sns.set_theme(style="whitegrid")

# 3. Vẽ biểu đồ Cột (Bar Chart) cho điểm trung bình
plt.figure(figsize=(10, 6))
sns.barplot(x=averages.index, y=averages.values, palette='viridis')

# Thêm nhãn điểm số trên đầu mỗi cột
for i, v in enumerate(averages.values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold', fontsize=12)

plt.title('Hiệu năng Hệ thống Hybrid RAG Y tế', fontsize=16)
plt.ylabel('Điểm số (0.0 - 1.0)', fontsize=12)
plt.ylim(0, 1.1)
plt.tight_layout()
plt.savefig('rag_metrics_averages.png') # Lưu biểu đồ cột
plt.show()

# 4. Vẽ biểu đồ Hộp (Box Plot) để xem độ ổn định (phân bổ điểm)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_revised[metrics], palette='Set2')
plt.title('Phân bổ điểm số của các chỉ số (Độ ổn định)', fontsize=16)
plt.ylabel('Điểm số', fontsize=12)
plt.tight_layout()
plt.savefig('rag_metrics_distribution.png') # Lưu biểu đồ phân bổ
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()
for i, metric in enumerate(metrics):
    sns.histplot(data_revised[metric], bins=15, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Phân bổ điểm: {metric.replace("_", " ").capitalize()}')

plt.tight_layout()
plt.savefig('rag_metrics_histograms.png')

# 2. Vẽ Biểu đồ Biến thiên (Trend)
plt.figure(figsize=(14, 6))
for metric in metrics:
    plt.plot(data_revised.index, data_revised[metric], label=metric, alpha=0.7)
plt.title('Biến thiên chất lượng qua 100 mẫu thử nghiệm')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('rag_metrics_trend.png')
```
